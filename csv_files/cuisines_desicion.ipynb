```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('HSR Layout.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
#hsr
food = []
i=0

while i<df.shape[0]:
    for x in df['dish'][i].split(','):
        food.append(x.strip().lower())
    i+=1

food = np.array(food)
hsr = np.unique(food)
print(len(hsr))
hsr
```

```python
menu = ['Pizza']
j=0
dfindex = []

while j<df.shape[0]:
    for x in menu:
        if x.lower() in df['dish'][j].lower():
            dfindex.append(j)
            break
    j+=1
```

```python
dffinal = df.iloc[dfindex]
```
